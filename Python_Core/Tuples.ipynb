# Tuples

#### Tuples are indexed , ordered (the order does not change) , immutable , can have duplicates 

```python
cars = ('bmw', 'mercedes', 'audi')
print(type(cars))
print(cars)

#indexed 
print(cars[0])
#immutable
# cars[0] = 'Fiat'   will generate an error
cars = ('bmw', 'mercedes', 'audi', 'mercedes')
print(cars)
# tuple with one element (need a comma)
car = ('bmw',)
# lenght
print(len(cars))

#tuple constructor  (pass any iterable)
words = tuple(['hello', 'ismail'])
print(words)
words = tuple(('hello', 'ismail', 'test'))
print(words)
dct = dict(first_name = 'ismail', last_name = 'yahyaoui')
word_keys = tuple(dct.keys())
word_values = tuple(dct.values())
word_items = tuple(dct.items())
print(word_keys)
print(word_values)
print(word_items)
```

```python
info = ('ismail', 'yahyaoui', 26, True)
# access with positive index
print(info[0])
# access with negative index
print(info[-1])
# access with slice  [start_idx : end_idx : step] end not part of the slice
print(info[1:3])
print(info[1:-2:1]) # can mix positive and negative , just need to follow the direction otherwise it will be empty
print(info[1:2])  # accessing with a slice returns a tuple even if the output is only one signgle element
print(info[::])
print(info[::-1])
```

```python
# since they are immutable , we convert the tuple to a list , do the changes , then change it back 
# Add
info = ('ismail', 'yahyaoui', 25)
print(info)
print(id(info))
tmp = list(info)
tmp.append(True)
info = tuple(tmp)
print(info)
print(id(info))  # the ids are obviously different because they are immutable a change is a new creation


# Update
print('*****************')
info = ('ismail', 'yahyaoui', 25)
print(info)
tmp = list(info)
tmp[0] = 'Hamid'
info = tuple(tmp)
print(info)

# Remove
print('*****************')
info = ('ismail', 'yahyaoui', 25)
print(info)
tmp = list(info)
tmp.pop()
info = tuple(tmp)
print(info)

#delete the entire tuple
print('*****************')
info = ('ismail', 'yahyaoui', 25)
del info
```

```python
# Unpacking the tuple
info = ('ismail', 'yahyaoui', 25)
first_name, last_name, age = info
print(first_name)
print(last_name)
print(age)

# Using * (spalt)

print("*****************")
cars = ('bmw', 'mercedes', 'audi', 'toyota', 'fiat')
car1 , car2, *car3 = cars
print(car1)
print(car2)
print(car3) # all the rest in a list 

print("*****************")
cars = ('bmw', 'mercedes', 'audi', 'toyota', 'fiat', 'renault')
car1 , *car2, car3 , car4 = cars
print(car1)
print(car2) # all the values until there only enough for the rest of the variables
print(car3) 
print(car4)

```

```python
test = ('test', 1, True)
print(test)
```

```python
# tuple
nums = (1, 2, 3)
print(*nums)

# list
nums = [1, 2, 3]
print(*nums)

# string  
letters = "abc"
print(*letters)   # same as f("a", "b", "c")

# set
nums = {1, 2, 3}
print(*nums)

# dict  (unpacks keys only)
d = {"a": 1, "b": 2}
print(*d)   # same as f("a", "b")
```
