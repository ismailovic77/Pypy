## Sets

##### set are unordered , unindexed , no duplicates , mutable

##### use sets for no duplication/uniqueness of data , 
##### fast lookup (works inside with hash , 0(1) operation to check is an smth in a set) and it's 0(n) for lists  
##### use also for math operation (intersection , union , difference)
##### The moment you catch yourself writing if x in some_list inside a loop, that's usually a sign a set would be better.

```python
utensils = {"spoon", "knife", "fork"}
print(utensils) # the print might give a different order each time than the order that we defined the set with => unordered

for _ in utensils:
    print(_)

# no dups
utensils = {"spoon", "knife", "fork", "fork", "fork"}
print(utensils)

```

```python
# Methods for sets
utensils = {"spoon", "knife", "fork"}
utensils.add(25)
print(utensils)
utensils.remove("fork")
print(utensils)
utensils.clear()
print(utensils)

del utensils
#print(utensils) this will result in a not Found error 
```

```python
utensils = {"spoon", "knife", "fork"}
dishes = ["bowl", "plate", "cup"]
utensils.update(dishes)
print(utensils)


utensils = {"spoon", "knife", "fork"}
dishes = ["bowl", "plate", "cup"]
table = utensils.union(dishes)  # this returns a new set 
print(table)

```

```python
## Comparaison
utensils = {"spoon", "knife", "fork"}
dishes = ["bowl", "plate", "cup", "spoon"]

diff = utensils.difference(dishes)
print(diff)

intersect = utensils.intersection(dishes)
print(intersect)
```
