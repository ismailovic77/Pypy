### Functions

```python
# function with normal params 
def add_number(op1, op2):
    return op1 + op2

val = add_number(5 , 6)
print(val)
```

```python
# function with default param
def add_number(op1, op2 = 6):
    return op1 + op2

val = add_number(2)
print(val)
val = add_number(1,2)
print(val)
```

```python
# function with *args (positional arguments)
def add_number(*args):
    return args[0] + args[1]

# when we define a functions with this type of params *args , the function expects to receive a comma separated values passed as params when 
# calling the fucntion like this add_numbers(1,2,3) , the function then takes those params into one single tuple that can be used to reference
# sigle params one by one

#callign the function 

#direct call 
res = add_number(1,2,4)
print(res)
# call with iterable inpacking 
numbers = (1,5,4)  # before passing this tuple we use the * to change into positional agruments
res = add_number(*numbers)
print(res)

numbers = [1,8,4]  # before passing this tuple we use the * to change into positional arguments
res = add_number(*numbers)
print(res)

```

```python
# function with **args 
def add_number(**args):
    return args['mark1'] + args['mark2']

# when we define a functions with this type of params **args , the function expects to receive multiple key value pairs passed as params ,
#  so when calling the fucntion like this add_numbers(op1=20,op2=30) , the function then takes those params into one single dict 
# that can be used to reference sigle params one by one

#callign the function 

#direct call 
res = add_number(mark1 = 10, mark2 = 20)
print(res)

# call with dictionary
numbers = dict(mark1 = 10, mark2 = 30)  # before passing this dict we use the ** to change into keyword arguments
res = add_number(**numbers)
print(res)
```

### Note

If you know exactly how many arguments you'll receive → just use normal params: (name, age)
If you don't know the number of params you'll receive , then you can use the *args or **args , depending on weather they are positional args 
or key value pairs arguments
