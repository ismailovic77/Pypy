## Lists

##### lists are mutable 

```python
# Adding elements to a list 
# append to add elements at the end of a list
lst = [1,2,3,4,5]
print(lst)
lst.append('ismail')
print(lst)
lst.append([34.44, True])
print(lst)

print('***************')

# insert() add elements at a certain position
lst.insert(0,"Hello")
print(lst)
lst.insert(2, False)
print(lst)
print('***************')

#extend() to add items of one list to another
lst1 = [1, 2, 3, 4]
lst2 = [5, 6, 7, 8]

lst1.extend(lst2)
print(lst1)
print(lst2)
```

```python
# input to a list
numbers = []
for i in range(4):
    x = int(input(f"enter the element {i}"))
    numbers.append(x)
print(numbers)

numbers = input("enter the numbers with a space : ").split()  #split(separator, maxsplit) default sep space
print(numbers)
```

```python
# update/insert elements on a list
test = ['ismail', "hello", 1, 43, True]
test[1] = ["yahyaoui", 34] # replace the value in index 1 with the whole list , no upacking 
print(test)
print("*************")
test[1:2] = ["yahyaoui", 34]  # inpacks the iterable and adds its elements in the specified range (replacing)
print(test)
print("*************")
```
