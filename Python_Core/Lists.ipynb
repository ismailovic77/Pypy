## Lists

##### lists are mutable , can have duplicates , ordered , indexed 

```python
# Adding elements to a list 
# append to add elements at the end of a list
lst = [1,2,3,4,5]
print(lst)
lst.append('ismail')
print(lst)
lst.append([34.44, True])
print(lst)

print('***************')

# insert() add elements at a certain position
lst.insert(0,"Hello")
print(lst)
lst.insert(2, False)
print(lst)
print('***************')

#extend() to add items of one list to another
lst1 = [1, 2, 3, 4]
lst2 = [5, 6, 7, 8]

lst1.extend(lst2)
print(lst1)
print(lst2)
```

```python
# input to a list
numbers = []
for i in range(4):
    x = int(input(f"enter the element {i}"))
    numbers.append(x)
print(numbers)

numbers = input("enter the numbers with a space : ").split()  #split(separator, maxsplit) default sep space
print(numbers)
```

```python
# update/insert elements on a list
test = ['ismail', "hello", 1, 43, True]
test[1] = ["yahyaoui", 34] # replace the value in index 1 with the whole list , no upacking 
print(test)
print("*************")
test[1:2] = ["yahyaoui", 34]  # inpacks the iterable and adds its elements in the specified range (replacing)
print(test)
print("*************")
```

```python
# removing items from a list

lst = ['ismail', "hello", 1, 43, True]
lst.remove('ismail') # removes using the item
print(lst)
print("************")

var = lst.pop(2) # removes based on index and returns the item removed
print(var)
print(lst)
lst.pop() # removes the last element of the lsit
print(lst)
print("************")

lst = ['ismail', "hello", 1, 43, True]
del lst[1]
print(lst)
del lst
# print(lst) this will return undefined  
print("*************")

lst = ['ismail', "hello", 1, 43, True]
lst.clear()
print(lst)
```

### List conprehension

```python
names = ["ismail", "yahyaoui", "Ilyasse", "Nizar"]
i_I_names = [name for name in names if name.lower().startswith('i')]
print(i_I_names)
i_names = [name for name in names if name.startswith('i')]
print(i_names)
```
