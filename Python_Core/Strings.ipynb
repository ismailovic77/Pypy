## Strings

```python
statment1 = "this is a string enclosed in double quotes"
statment2 = 'this is a string enclosed in single quotes'
print(statment1)
print(statment2)

statment3 = "this is a string with 'single quotes' inside" #should be enclosed in double quotes
statment4 = 'this is a string with "double quotes" inside' #should be enclosed in single quotes
statment5 = 'this is a string with an \'escape' # we can always use an escape character
print(statment3)
print(statment4)
print(statment5)

#multiline
statment6 = """this is a multiline string
we can write stuff on multiple lines
"""
print(statment6)
```

```python
# Membership operators
test = "jas" in "jasmine"
print(test)

test = "jas" not in "jasmine"
print(test)

#Escape characters : 
    #\n new line
    #\b backspace
    #\t tab
    #\ normal escape 
    #\\ escape the escape

# Slicing  
# s[i:j:k] , i is the start index , j is the stopping index but the value in it is not included , k step (could be negative for reverse parsing)
var = "jasmine"
print(var[::]) # from end to end with default step 1
print(var[:4:2])
print(var[::-1]) # to reverse a string 
```

#### String Formatting (old)

```python
#simple % formating
first_name = "ismail"
last_name = "yahyaoui"
str = "Hello world %s" % first_name
print(str)
print("************")

# multiple variable % formating
str = "Hello world %s %s" %(first_name, last_name)
print(str)
print("************")

#simple str.format() => use the {}
s = "My first name is {} and my last name is {}".format(first_name, last_name)
print(s)
print("************")

#str format with indexing
s = "My first name is {0} and my last name is {1}".format(first_name, last_name)
print(s)
s = "My first name is {0} and my last name is {0}".format(first_name, last_name) # playing with the indexes
print(s)
print("************")

#str format with varbiable = 
s = "My first name is {FirstName} and my last name is {LastName}".format(FirstName=first_name,LastName=last_name)
print(s)
print("************")

#str format with format specifiers (float , binary , integrer , string) + example with casting , and finallly control of the decimal
s = "I got {0:f} in the exam".format(55.66)  # 0 here is the index
print(s)
s = "I got {0:f} in exam and {1:f} in test".format(55,66.66)  #with INT # with a string it gives an error (no typecating for float to str)
print(s)
s = "I got {0:.2f} in exam and {1:.3f} in test".format(55,66.6645)  #with INT # with a string it gives an error (no typecating for float to str)
print(s)
```

#### String Formatting (new)

```python


# evaluation in curly braces and escaping them
# f string formatting
city = "kenitra"
country = "Morocco"
profession = "Data Engineer"

str = f"I live in {city}, {country}"
print(str)
print("************")

# f string with method call
str = f"I live in {city.capitalize()}, {country.upper()}"
print(str)
print("************")

# multiline commands with backslash , () , """
str = f"I live in {city} "\
    f"I work in the country {country} " \
    f"I work as a {profession}"
print(str)

str = (f"I live in {city} "
    f"I work in the country {country} " 
    f"I work as a {profession}")
print(str)

str = f"""I live in {city}
    I work in the country {country} 
    I work as a {profession}"""
print(str)  # with this we keep the tabs the newline and everything added to the string
print("************")

#escaping curly braces , and evaluation
x = 10
y = 20
str = f"the sum x+y is equal to {x+y}" #evaluation
print(str)

str = f"The sum of {{x+y}} is equal to {x+y}" #escaping
print(str)

str = f"The sum of {{x+y}} is equal to {{{x+y}}}" #escaping and evaluation
print(str)

# the general format for string formatting here if you wanna play with data types is this : 
# {value : [fill][align][sign][width][.precision][type]}
```
