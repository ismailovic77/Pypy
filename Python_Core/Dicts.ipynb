# Dictionaries

```python
# the creation of a dictionary
car = {'brand': 'Audi', 'model': 'q7'}
print(car)
# no duplicates
car = {'brand': 'Audi', 'model': 'q7', 'model': 'q8'}
print(car)
# mutability => dicts are mutable , changind values are happening on the same object , keeps the same id after changing values
car['model'] = 'q9'
print(car)
# lenght of dict
print(len(car))
# dict() consutructor
car = dict(brand = 'Audi', model = 'V8')
print(car)

```

```python
car = {'brand': 'Audi', 'model': 'V8'}

# accessing value with key name
print(car['brand'])
print('****************')
# accssing values using the get method() (there is another get method I thnk )
print(car.get('model'))
print('****************')
# accessing keys using the keys() / view objects
car_keys = car.keys() # this is a veiw object , tracks changes automatically 
print(car_keys)
car['gaz'] = 'petrol'
print(car_keys)
print('****************')
# accessing values using the values() / view objects
car_values = car.values() # this is a veiw object , tracks changes automatically 
print(car_values)
car['test'] = True
print(car_values)
print('****************')
# accessing items using the items // view objects
car_items = car.items() # this is a veiw object , tracks changes automatically 
print(car_items)
car['motor'] = 'Good'
print(car_items)
```

```python
# Update/add value using the update method
car = dict(brand = 'Audi', model = 'V6')
print(id(car))
print(car)
car.update({'model': 'Q7'})
print(id(car))
print(car)
car.update({'gaz': 'petrol'})
print(car)
```

```python
# remove item with pop()
car = dict(brand = 'Audi', model = 'V6')
it = car.pop('model') # returns the value of the deleted item
print(car)
print(it) 
print('****************')
# remove item with popitem()
car = dict(brand = 'Audi', model = 'V6')
car['gaz'] = 'petrol'
it = car.popitem()  # removes the latest added item and returns the deleted item as a tuple
print(car)
print(it)
print('****************')
# remove item with del
car = dict(brand = 'Audi', model = 'V6')
del car['brand']
print(car)
print('****************')
# remove dict with del
car = dict(brand = 'Audi', model = 'V6')
del car
#print(car) returns an undefined exception
# clear with with clear()
car = dict(brand = 'Audi', model = 'V6')
car.clear()
print(car)
```

```python
"""
Mutability & Copy in Python
Memory Model: Labels & Objects
When you write x = 10, Python allocates a space in memory holding the value 10, and x is simply a label (reference) pointing to that space — not the value itself.

Immutable Objects (e.g. int, str, tuple)
pythonx = 10
y = x        # y points to the same object as x
id(x) == id(y)  # True

y = 20       # creates a NEW object in memory; x is untouched
Because integers are immutable, Python knows that reassigning y can never affect x — so it's safe to let them share the same object initially.

Optimization note: For small integers (and equal immutables like x = 10; y = 10), Python only creates one object and points both labels at it. Since neither can mutate it, there's no risk.


Mutable Objects (e.g. dict, list)
pythonx = {'model': 'mercedes'}
y = x          # y points to the SAME object
y['model'] = 'bmw'
print(x)       # {'model': 'bmw'} — x changed too!
Because both labels point to the same object in memory, mutating through y also affects x.
This is why Python creates separate objects for two dicts with the same value:
pythonx = {'model': 'mercedes'}
y = {'model': 'mercedes'}
id(x) == id(y)  # False — two distinct objects
Unlike small integers, Python can't optimize these away — a dict can be mutated, so sharing would be unsafe.

Making a True Copy
When you want an independent copy of a mutable object:
python# Shallow copy options
y = x.copy()
y = dict(x)
y = x[:]        # for lists

# Deep copy (nested structures)
import copy
y = copy.deepcopy(x)
Now x and y point to different objects — mutating one leaves the other untouched.

"""
```

```python
# create a copy using copy()
car = dict(brand = 'Audi', model = 'V6')
car_copy = car.copy()
print(car)
print(car_copy)
car_copy['model'] = 'v8'
print(car)
print(car_copy)
print('***********')
# create a copy using dict()
car = dict(brand = 'Audi', model = 'V6')
car_copy = dict(car)
print(car)
print(car_copy)
car_copy['model'] = 'v8'
print(car)
print(car_copy)
print('***********')
```
